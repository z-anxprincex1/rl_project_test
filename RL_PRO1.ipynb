```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
asset = "BTC-USD"
start_date = "2008-01-01"
end_date = "2023-01-01"
interval = "1d"

btc_usd = yf.download(asset, start=start_date, end=end_date, interval=interval).reset_index()
btc_usd.to_csv("btc_usd.csv", index = False)
```

```python
df2 = df1.copy()
```

```python
df2['Close'] = pd.to_numeric(df2['Close'], errors='coerce')
```

```python
def calculate_macd(df, short_window=12, long_window=26, signal_window=9):
    short_ema = df['Close'].ewm(span=short_window, adjust=False).mean()
    long_ema = df['Close'].ewm(span=long_window, adjust=False).mean()
    macd = short_ema - long_ema
    signal = macd.ewm(span=signal_window, adjust=False).mean()
    return macd, signal

df2['MACD'], df2['Signal Line'] = calculate_macd(df2)
```

```python
df2['MACD'].describe()
```

```python
def calculate_ema(df, span):
    ema = df['Close'].ewm(span=span, adjust=False).mean()
    return ema

df2['EMA'] = calculate_ema(df2, span=20)
```

```python
import numpy as np
import pandas as pd

class TradingEnvironment:
  def __init__(self, data):
    self.data = data
    self.current_step = 0
    self.total_steps = len(data) - 1
    self.initial_balance = 10000
    self.balance = self.initial_balance
    self.shares_held = 0
    self.net_worth = self.initial_balance
    self.prev_net_worth = self.initial_balance

  def reset(self):
    self.current_step = 0
    self.balance = self.initial_balance
    self.shares_held = 0
    self.net_worth = self.initial_balance
    return self._get_state()

  def _get_state(self):
    state = [
      self.data['Close'][self.current_step],
      self.data['MACD'][self.current_step],
      self.data['Signal Line'][self.current_step],
      self.data['EMA'][self.current_step],
      self.balance,
      self.shares_held,
      self.net_worth,
    ]
    return np.array(state)

  def step(self, action):
    current_price = self.data['Close'][self.current_step]

    if action == 0:
      if buy_signal(self.data.iloc[self.current_step]):
        shares_to_buy = int(self.balance / current_price)
        if shares_to_buy > 0:
            cost = shares_to_buy * current_price
            self.balance -= cost
            self.shares_held += shares_to_buy

    elif action == 1:
      if sell_signal(self.data.iloc[self.current_step]):
        if self.shares_held > 0:
            revenue = self.shares_held * current_price
            self.balance += revenue
            self.shares_held = 0

    self.net_worth = self.balance + (self.shares_held * current_price)
    reward = self.net_worth - self.prev_net_worth
    self.prev_net_worth = self.net_worth
    done = (self.current_step == self.total_steps)

    if not done:
        self.current_step += 1

    return self._get_state(), reward, done, self.balance, self.net_worth
```
